#Define Function

```python
#function to check Refernce URL
def FindUrl(inString):
    if "https" in inString :
      return 1
    else:
      return 0
```

```python
#function to check age
def AgeCheck(age):
  if age >21:
    return 0
  else:
    return 1
```

```python
LangPatterns =['girls','models','students','provide premium service','24X7','24/7','call anytime','young' ,'tender','escorts','vip']
```

```python
def LangPattern(word, lists):
    x=word.split()
    for i in x:
      if i in lists:
          return 1
          break
    return 0
```

```python
def Trafficked(Age,URL, Lang):
  if Age ==1 and URL ==1 and Lang ==1:
    return 1
  else:
    return 0
```

# **DATA CLEANING & PREPARATION**

```python
# Drops the columns which are not required for the analysis
df.columns
df.drop(['url', 'date_posted'],axis=1,inplace=True)
```

```python
#Clean description/Title cloumn by remove the escape characters
df['description'] = df['description'].replace(to_replace= '\W', value= ' ', regex=True)
df['Title'] = df['Title'].replace(to_replace= '\W', value= ' ', regex=True)
df.head(3)
```

```python
# Drop all observations with all attributes as NaN
df = df.dropna(how='any',axis=0)
df.head(3)
```

```python
df.isna().sum()
```

```python
df['websiteRef']=[FindUrl(x) for x in df['description']]
```

```python
df['websiteRef'].value_counts()
```

```python
df['Minor']=[AgeCheck(x) for x in df['Age']]
```

```python
df['Minor'].value_counts()
```

```python
df['LangPattern']=[LangPattern(x.lower(),LangPatterns) for x in df['Title']]
```

```python
df['LangPattern'].value_counts()
```

```python
res=[]
for index, row in df.iterrows():
  res.append(Trafficked(row['Minor'], row['websiteRef'],row['LangPattern']))

df['Trafficked']=res
```

```python
df['Trafficked'].value_counts()
```

# Cleaning the '**Description**'

```python
#Convert the string into lowercase
df['description'] = df['description'].apply(lambda x: x.lower())
print(df['description'])
```

```python
#remove punctuation
df['description']  = df['description'].apply(lambda x: re.sub('[%s]' % re.escape(string.punctuation), '' , x))
print(df['description'] )
```

```python
df['description'] = df['description'].apply(lambda x: re.sub('[0-9]','',x))
print(df['description'] )
```

# **Cleaning 'Title'**

```python
#Convert the string into lowercase
df['Title'] = df['Title'].apply(lambda x: x.lower())
print(df['Title'])
```

```python
#remove punctuation
df['Title']  = df['Title'].apply(lambda x: re.sub('[%s]' % re.escape(string.punctuation), '' , x))
print(df['Title'] )
```
